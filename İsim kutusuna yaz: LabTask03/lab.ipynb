```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")

df = pd.read_csv('/content/drive/MyDrive/IAT461_LAB03/music_streams.csv')
print(df.head())
```

```python
!pip install altair
import altair as alt
```



## Avarage Streams by Year


```python

yearly = df.groupby('year')['streams_millions'].mean().reset_index()

chart = alt.Chart(yearly).mark_bar().encode(
    x=alt.X('year:O', title='Year'),
    y=alt.Y('streams_millions:Q', title='Average Streams (M)'),
    color=alt.Color('streams_millions:Q', scale=alt.Scale(scheme='blues'))
).properties(
    title='Average Monthly Streams by Year',
    width=400,
    height=200
)

chart
```

##Tempo Trend Over Years

```python
tempo_trend = df.groupby('year')['tempo_bpm'].mean().reset_index()

chart2 = alt.Chart(tempo_trend).mark_line(point=True, color='crimson').encode(
    x=alt.X('year:O', title='Year'),
    y=alt.Y('tempo_bpm:Q', title='Average Tempo (BPM)', scale=alt.Scale(zero=False)),
    tooltip=['year', 'tempo_bpm']
).properties(
    title='Average Song Tempo Over the Years',
    width=600,
    height=300
)

chart2
```

##Genre popularity over years

```python

genre_trend = df.groupby(['year', 'genre'])['streams_millions'].mean().reset_index()

chart3 = alt.Chart(genre_trend).mark_line(point=True).encode(
    x=alt.X('year:O', title='Year'),
    y=alt.Y('streams_millions:Q', title='Average Streams (M)'),
    color=alt.Color('genre:N', title='Genre'),
    tooltip=['year', 'genre', 'streams_millions']
).properties(
    title='Genre Popularity Over the Years',
    width=600,
    height=300
)

chart3
```
